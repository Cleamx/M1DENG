```python
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
import numpy as np

img = Image.open("img/image.png")
plt.imshow(img)
```

```python
imgNp = np.array(img.convert("L"))
Longueur, Largeur = imgNp.shape
plt.imshow(imgNp)
```

```python
noyau_vertical = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
noyau_horizontal = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

imgConvolution = np.copy(imgNp)

for line in range(1, Longueur-1):
    for col in range(1, Largeur-1):
        somme_v, somme_h = 0, 0
        for l in range(3):
            for c in range(3):
                somme_v += imgNp[line+l-1, col+c-1] * noyau_vertical[l, c]
                somme_h += imgNp[line+l-1, col+c-1] * noyau_horizontal[l, c]
        imgConvolution[line, col] = (somme_v * somme_v + somme_h * somme_h) ** 0.5
```

```python
plt.imshow(imgConvolution)
```
